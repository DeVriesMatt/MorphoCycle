```python
from pathlib import Path

fluor = Path("/media/mvries/Derek_Jeeters/PCNA_cell_cycle_marker/data_analysis/mCh_100Pc_subBkg/")
qpi = Path("/media/mvries/Derek_Jeeters/PCNA_cell_cycle_marker/data_analysis/reconstructed_QPI_registered/")

fluor_save = Path("/mnt/nvme0n1/Datasets/PCNA/PhaseToFluor/mCh_100Pc_subBkg/time_split/")
qpi_save = Path("/mnt/nvme0n1/Datasets/PCNA/PhaseToFluor/reconstructed_QPI_registered/time_split/")
```

```python
import tifffile as tfl
from tqdm import tqdm

for file in tqdm(qpi.iterdir()):
    image = tfl.imread(file)
    for t in tqdm(range(image.shape[0])):
        (qpi_save / file.stem).mkdir(parents=True, exist_ok=True)
        
        tfl.imwrite(qpi_save / file.stem / f"{str(t).zfill(5)}.tif",image[t])
```
