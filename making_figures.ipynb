```python
import numpy as np

def rescale_image(img, old_min, old_max, new_min, new_max):
    return ((img - old_min) / (old_max - old_min)) * (new_max - new_min) + new_min

# Example usage:
rescaled_img = rescale_image(img, img.min(), img.max(), 0, 255)

print(rescaled_img)
```

```python
import skimage
skimage.io.imsave("g1.png", rescaled_img)
```
